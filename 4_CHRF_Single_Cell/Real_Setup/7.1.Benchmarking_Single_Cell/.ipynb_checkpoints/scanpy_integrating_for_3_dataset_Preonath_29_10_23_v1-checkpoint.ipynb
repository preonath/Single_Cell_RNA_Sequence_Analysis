```python
import scanpy as sc
import os
```

```python
# Specify the directory containing your .h5ad files
directory = './data/merged_3_dataset/Longitudinal_profiling_of_respiratory_and_systemic_immune_responses_reveals_myeloid_cell_driven_lung_inflammation_in_severe_COVID_19/'

# List all .h5ad files in the directory and remove invisible characters
h5ad_files = [f.replace('\u200b', '') for f in os.listdir(directory) if f.endswith('.h5ad')]

# Read each .h5ad file and store the AnnData objects in a list
adata_list = [sc.read_h5ad(os.path.join(directory, file)) for file in h5ad_files]

# Concatenate all AnnData objects into a single object using anndata.concat
adata_1 = sc.concat(adata_list, join='outer', index_unique='-')

display(adata_1)
```

```python
from Bio import Entrez
import pandas as pd

Entrez.email = "YOUR_EMAIL"

def get_gene_name(ensembl_id):
    try:
        handle = Entrez.esearch(db="gene", term=ensembl_id)
        record = Entrez.read(handle)
        if not record["IdList"]:
            return None
        gene_id = record["IdList"][0]
        handle = Entrez.efetch(db="gene", id=gene_id, retmode="xml")
        record = Entrez.read(handle)
        gene_name = record[0]["Entrezgene_gene"]["Gene-ref"]["Gene-ref_locus"]
        return gene_name
    except Exception as e:
        print(f"Could not retrieve gene name for {ensembl_id}: {e}")
        return None

# Let's fetch gene names for the first 10 Ensembl IDs as an example
ensembl_ids = adata_1.var_names[:10]
gene_names = [get_gene_name(ensembl_id) for ensembl_id in ensembl_ids]

# Create a DataFrame to store the results
gene_info = pd.DataFrame({
    "Ensembl_ID": ensembl_ids,
    "Gene_Name": gene_names
})

print(gene_info)
```

```python
import scanpy as sc
import pandas as pd

# Assuming adatas is a list containing your loaded datasets
# Concatenate your annotated datasets
adata = adatas[0].concatenate(adatas[1:], batch_key='batch', batch_categories=['Batch_1', 'Batch_2', 'Batch_3'])

# Step 1: Data Preprocessing
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata = adata[:, adata.var['highly_variable']]
sc.pp.scale(adata, max_value=10)

# Step 2: Batch Correction
# Note: Combat might not be the best choice for every dataset. Always explore alternative batch correction methods.
sc.pp.combat(adata, key='batch')

# Step 3: Dimensionality Reduction
sc.tl.pca(adata, svd_solver='arpack')

# Compute the neighborhood graph
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)  # You need to adjust n_neighbors and n_pcs according to your data

# UMAP
sc.tl.umap(adata)

# Step 4: Clustering
sc.tl.leiden(adata)

# Step 5: Differential Expression Analysis
sc.tl.rank_genes_groups(adata, 'leiden', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=25, sharey=False)

# Visualization: UMAP plot
sc.pl.umap(adata, color=['leiden', 'batch'])

# Show top 10 genes per cluster
result = adata.uns['rank_genes_groups']
groups = result['names'].dtype.names
pd.DataFrame({group + '_' + key[:1]: result[key][group]
    for group in groups for key in ['names', 'pvals']}).head(10)
```

```python
import scanorama

# Only retain the highly-variable genes identified in the individual datasets
for adata in adatas:
    adata = adata[:, adata.var['highly_variable']]

# Integration
adatas_corrected = scanorama.correct_scanpy(adatas, return_dimred=True)

# Concatenation of the corrected datasets
adata_combined = adatas_corrected[0].concatenate(adatas_corrected[1:])

```

```python
# Dimensionality Reduction
sc.tl.pca(adata_combined, svd_solver='arpack')
sc.pp.neighbors(adata_combined)
sc.tl.umap(adata_combined)

# Clustering
sc.tl.leiden(adata_combined)

# Visualization
sc.pl.umap(adata_combined, color=['leiden', 'batch'])

```
